```python
import requests
from bs4 import BeautifulSoup
from time import sleep

```

```python
data = []

for p in range(1, 3):
    print(p)

    url = "https://www.toy.ru/catalog/producers/L.O.L./?PAGEN_7={p}"
    r = requests.get(url)

    soup = BeautifulSoup(r.text, 'lxml')

    dolls = soup.findAll('div', class_='col-12 col-sm-6 col-md-6 col-lg-4 col-xl-4 my-2')

    

    for doll in dolls:
        link = 'LINK: https://www.toy.ru'+doll.find('a', class_='d-block img-link text-center gtm-click').get('href')+' '
        doll_name = 'NAME:'+doll.find('a', class_='d-block p-1 product-name gtm-click').text
        doll_brand = 'BRAND:'+doll.find('a', class_='d-block p-1 product-name gtm-click').get('gtm-brand')
        doll_id = 'ID:'+doll.find('a', class_='d-block p-1 product-name gtm-click').get('gtm-id')
        doll_price = 'PRICE:'+doll.find('div', class_='col-12 align-self-end').find('span', class_='price').text
    
        data.append([link, doll_name, doll_brand, doll_id, doll_price])
    
data
```
